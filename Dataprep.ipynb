# Manipulando o banco de dados

A partir do JSON salvo, podemos extrair qualquer informação que desejarmos.

```python
import json
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
with open('responses.json','r') as json_file:
     db = json.load(json_file)
```

# Palavras mais comuns

Exemplo de manipulação, ao gerar uma nuvem de palavras com base nos tokens.

```python
# Filtrando apenas introdução
lessons_intro = [
     lesson
     for lesson in db
     if lesson['skillId'] == 'c796eab5a72dd5c83670aa63876020c5'
]

def generate_word_cloud(lessons):

    tokens = []

    for lesson in lessons:
        for challenge in lesson['challenges']:
            tokens = tokens + challenge.get('correctTokens', [])

    wc = WordCloud(background_color="white")
    wc.generate(" ".join(tokens))

    plt.figure()
    plt.imshow(wc) 
    plt.axis("off")

generate_word_cloud(lessons_intro)
```
